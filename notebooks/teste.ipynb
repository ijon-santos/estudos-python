```python
# Teste TensorFlow (se instalado)
try:
    import tensorflow as tf
    print(f"✅ TensorFlow version: {tf.__version__}")
    print(f"   GPUs disponíveis: {tf.config.list_physical_devices('GPU')}")
except ImportError:
    print("⚠️ TensorFlow não está instalado (OK, é opcional)")
```

```python
# Teste visualização simples
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='seno', color='blue', linewidth=2)
plt.title('Gráfico de Teste')
plt.xlabel('x')
plt.ylabel('seno(x)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
