# 📘 Python - Erros e Exceções

## 🎯 Objetivos
- Entender a diferença entre erros e exceções
- Tratar exceções com try/except
- Levantar exceções com raise
- Criar exceções personalizadas

## 📝 Conteúdo

### 1. Tipos de Erros
- Erros de sintaxe (SyntaxError)
- Exceções (NameError, TypeError, ValueError, etc.)

### 2. Tratamento de Exceções
```python
try:
    # código que pode gerar erro
except TipoDeErro:
    # código para tratar o erro
else:
    # código se não houve erro
finally:
    # código executado sempre
```

### 3. Levantar Exceções
- `raise`
- Criar exceções personalizadas
