### 4. Unicode e Normalização
- Problemas com diferentes representações
- `unicodedata.normalize()`
- Formas: NFC, NFD, NFKC, NFKD
- Remoção de acentos para buscas

```python
import unicodedata
s = 'São Paulo'
nfd = unicodedata.normalize('NFD', s)
sem_acentos = ''.join(c for c in nfd if not unicodedata.combining(c))
print(sem_acentos)  # 'Sao Paulo'
```
