### 1.Estruturas de Controle

Um programa é formado por instruções. Sem estruturas de controle, essas instruções seriam executadas apenas:\
<br>
de cima para baixo;\
uma única vez;\
sempre na mesma ordem.\
<br>
As estruturas de controle permitem alterar esse fluxo.\
<br>
Elas respondem a perguntas como:\
<br>
Este bloco deve ser executado?\
Quantas vezes ele deve ser executado?\
Quando o processamento deve parar?\
O que fazer quando ocorre um erro?\
Como garantir a liberação de um recurso?\
Como suspender e retomar uma execução?\
Como escolher uma ação com base na estrutura dos dados?\
<br>
Em Python, as principais formas de controle são:\
<br>
| Categoria                  | Estruturas                                          |
| -------------------------- | --------------------------------------------------- |
| Decisão                    | `if`, `elif`, `else`                                |
| Repetição                  | `for`, `while`                                      |
| Interrupção                | `break`, `continue`, `return`, `raise`              |
| Tratamento de erros        | `try`, `except`, `else`, `finally`                  |
| Gerenciamento de contexto  | `with`, `async with`                                |
| Correspondência estrutural | `match`, `case`                                     |
| Execução preguiçosa        | `yield`, geradores                                  |
| Concorrência assíncrona    | `async`, `await`, `async for`                       |
| Expressões de controle     | `and`, `or`, `not`, expressão condicional           |
| Controle declarativo       | compreensões, `any`, `all`, despacho por dicionário |
<br>
A documentação da linguagem classifica if, while, for, try, with, match, funções, classes e estruturas assíncronas como compound statements, ou instruções compostas: instruções que controlam outras instruções.
<br>

### Regra prática
Use quatro espaços por nível:\
```
if condicao:
    instrucao_1()

    if outra_condicao:
        instrucao_2()
```
⚠️**Evite misturar tabulações e espaços.**⚠️\
<br>
Misturar tabulações (Tabs) e espaços no Python é um dos caminhos mais rápidos para conseguir uma dor de cabeça.\
A resposta curta é: o Python 3 simplesmente não permite isso e vai travar o seu código (gerando um erro chamado TabError).\
Para entender o porquê de forma simples e direta, veja os motivos principais:\

1. **Invisibilidade do problema**\
Visualmente, uma tabulação e um conjunto de 4 espaços podem parecer exatamente iguais no seu editor de texto. \
No entanto, para o interpretador do Python, eles são caracteres completamente diferentes:\
Espaço: Caractere ASCII 32.\
Tabulação (Tab): Caractere ASCII 9.\
Se você misturar os dois, o Python não conseguirá saber qual nível de indentação pertence a qual bloco de código.
<br>
2. **O TabError**\
Como o Python utiliza a indentação para definir onde começam e terminam as funções, condições (if) e laços de repetição (for/while), a consistência é vital.\
Se você tentar executar um código misto no Python 3, ele retornará imediatamente este erro:\
TabError: inconsistent use of tabs and spaces in indentation
<br>
3. **Código que mente para você**\
Em alguns editores de texto antigos ou mal configurados, um "Tab" pode ser renderizado visualmente como 8 espaços, enquanto em outros é renderizado como 4 ou até 2 espaços.\
Se você misturar os dois:\
O código pode parecer perfeitamente alinhado na sua tela.\
Ao rodar (ou abrir em outro computador), a lógica real do programa será interpretada de forma totalmente diferente da que você está vendo, gerando bugs silenciosos e difíceis de rastrear.\
<br>
**O Padrão Oficial (PEP 8)**\
A PEP 8, que é o guia de estilo oficial da linguagem Python, dita uma regra muito clara:\
Use apenas espaços para indentação.\
O padrão recomendado são 4 espaços por nível.\
<br>
Praticamente todo editor moderno (VS Code, PyCharm, Cursor, etc.) possui uma configuração para "converter tabs em espaços" automaticamente.\
Deixe essa opção ativada! Assim, você pode apertar a tecla Tab no teclado e o editor insere silenciosamente 4 espaços para você.\
