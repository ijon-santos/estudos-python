## 📝 Conteúdo

### 1. Estruturas Condicionais
```python
if condicao:
    # código
elif outra_condicao:
    # código
else:
    # código
```

### 2. Estruturas de Repetição
```python
for item in iteravel:
    # código

while condicao:
    # código
```

### 3. Break, Continue, Else
- `break`: interrompe o loop
- `continue`: pula para a próxima iteração
- `else`: executa se o loop não foi interrompido
