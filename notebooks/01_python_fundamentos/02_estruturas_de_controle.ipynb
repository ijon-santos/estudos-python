### 1.Estruturas de Controle

Um programa é formado por instruções. Sem estruturas de controle, essas instruções seriam executadas apenas:\
<br>
de cima para baixo;\
uma única vez;\
sempre na mesma ordem.\
<br>
As estruturas de controle permitem alterar esse fluxo.\
<br>
Elas respondem a perguntas como:\
<br>
Este bloco deve ser executado?\
Quantas vezes ele deve ser executado?\
Quando o processamento deve parar?\
O que fazer quando ocorre um erro?\
Como garantir a liberação de um recurso?\
Como suspender e retomar uma execução?\
Como escolher uma ação com base na estrutura dos dados?\

### 2.Blocos e Indentação

Use quatro espaços por nível:\
```
if condicao:
    instrucao_1()

    if outra_condicao:
        instrucao_2()
```
⚠️**Evite misturar tabulações e espaços.**⚠️\
<br>
Misturar tabulações (Tabs) e espaços no Python é um dos caminhos mais rápidos para conseguir uma dor de cabeça.\
A resposta curta é: o Python 3 simplesmente não permite isso e vai travar o seu código (gerando um erro chamado TabError).\
Para entender o porquê de forma simples e direta, veja os motivos principais:\

1. **Invisibilidade do problema**\
Visualmente, uma tabulação e um conjunto de 4 espaços podem parecer exatamente iguais no seu editor de texto. \
No entanto, para o interpretador do Python, eles são caracteres completamente diferentes:\
Espaço: Caractere ASCII 32.\
Tabulação (Tab): Caractere ASCII 9.\
Se você misturar os dois, o Python não conseguirá saber qual nível de indentação pertence a qual bloco de código.
2. **O TabError**\
Como o Python utiliza a indentação para definir onde começam e terminam as funções, condições (if) e laços de repetição (for/while), a consistência é vital.\
Se você tentar executar um código misto no Python 3, ele retornará imediatamente este erro:\
TabError: inconsistent use of tabs and spaces in indentation
3. **Código que mente para você**\
Em alguns editores de texto antigos ou mal configurados, um "Tab" pode ser renderizado visualmente como 8 espaços, enquanto em outros é renderizado como 4 ou até 2 espaços.\
Se você misturar os dois:\
O código pode parecer perfeitamente alinhado na sua tela.\
Ao rodar (ou abrir em outro computador), a lógica real do programa será interpretada de forma totalmente diferente da que você está vendo, gerando bugs silenciosos e difíceis de rastrear.
4. **O Padrão Oficial (PEP 8)**\
A PEP 8, que é o guia de estilo oficial da linguagem Python, dita uma regra muito clara:\
Use apenas espaços para indentação.\
O padrão recomendado são 4 espaços por nível.\

Praticamente todo editor moderno (VS Code, PyCharm, Cursor, etc.) possui uma configuração para "converter tabs em espaços" automaticamente.\
Deixe essa opção ativada! Assim, você pode apertar a tecla Tab no teclado e o editor insere silenciosamente 4 espaços para você.

### 3. Valores booleanos e “truthiness”
Uma condição não precisa produzir literalmente True ou False. Python interpreta muitos objetos como verdadeiros ou falsos.
```
nome = "Gustavo"

if nome:
    print("O nome foi preenchido")
```
A string não vazia é considerada verdadeira.

### Valores normalmente considerados falsos
São falsos em contexto booleano:
```
False
None
0
0.0
0j
""
[]
()
{}
set()
range(0)
```
Exemplos:

````python
valores = []

if not valores:
    print("A lista está vazia")
````

````python
resultado = None

if resultado is None:
    print("Ainda não existe resultado")
````

Objetos podem personalizar sua avaliação booleana por meio de `__bool__()` ou, na ausência dele, de `__len__()`.

#### Não confundir ausência com zero
Considere:
```
desconto = 0
```
Isto pode significar que existe um desconto válido de zero por cento.
```
if not desconto:
    print("Sem desconto")
```
O código mistura duas situações:

desconto is None: desconto não informado;
desconto == 0: desconto informado, mas igual a zero.

Mais preciso:

````python
desconto = 0
if desconto is None:
    print("Desconto não informado")
else:
    print(f"Desconto: {desconto}%")
````

#### Regra mais profissional

Use truthiness quando vazio e ausente tiverem o mesmo significado:

```
if not produtos:
    print("Nenhum produto")
```
Use comparação explícita quando os estados tiverem significados diferentes:
```
if quantidade is None:
    print("Quantidade desconhecida")
elif quantidade == 0:
    print("Produto esgotado")
```

### 4. Operadores booleanos

Python possui:

```
and
or
not
```
`AND` - A expressão é verdadeira quando todos os operandos são verdadeiros.\
`OR` - É verdadeira quando pelo menos um operando é verdadeiro.\
`NOT` - Inverte a interpretação booleana.\
exemplos:

````python
#ex1
dia = "sábado"

if dia == "sábado" or dia == "domingo":
    print("Fim de semana")
````

Uma forma melhor:

````python
#ex2
if dia in {"sábado", "domingo"}:
    print("Fim de semana")
````

Por que o ex2 é melhor?
Resposta: 
1. **Legibilidade e Concisão**\
   Você precisa repetir a variável dia e o operador de comparação == para cada item.\
   Se você quisesse verificar 5 dias úteis, a linha ficaria gigante e repetitiva.
2. **Escalabilidade (Fácil de manter)**
3. **Performance (Busca em Tempo Constante)**\
- Como funciona o `or`:\
  O Python avalia cada condição uma por uma, da esquerda para a direita (complexidade linear, ou $O(n)$).\
  Se você tiver 100 condições ligadas por or, ele fará até 100 comparações.
- Como funciona o `in` com `set`:\
  Os conjuntos em Python utilizam uma tabela Hash por baixo dos panos.\
  Isso significa que a busca de um elemento dentro de um set é feita em tempo constante ($O(1)$), \
  não importa se o conjunto tem 2 ou 2.000.000 de elementos. \
  O Python vai direto ao endereço de memória correto para checar se o elemento existe ali.
  
  ⚠️ Nota de rodapé: Para apenas 2 elementos, a diferença de velocidade é imperceptível.\
   Mas criar o hábito de usar `in` com `set` (ou mesmo com `list` ou `tuple` quando a ordem importar)\
  prepara seu código para lidar com grandes volumes de dados de forma eficiente.


#### Curto-circuito (short-circuit evaluation)
Funciona como uma "rede de segurança" para evitar que seu código quebre.

Pense no operador `and` como um detector de mentiras bem rígido.\
Para que uma expressão com `and` seja verdadeira, todas as partes dela precisam ser verdadeiras:\
$$\text{Verdadeiro} \text{ and } \text{Verdadeiro} \implies \text{Verdadeiro}$$

Se a primeira parte que o Python lê já for Falsa, o Python é "inteligente" (e preguiçoso).\
Ele pensa:"Se a primeira parte já é Falsa, não importa o que venha depois, \
o resultado final do `and` vai ser Falso de qualquer jeito.\
Não preciso perder tempo lendo o resto."Ele interrompe a leitura ali mesmo. Isso é o curto-circuito.

#### Se fosse com `or`?

O curto-circuito também acontece com o `or`, mas ao contrário.\
Para o or ser verdadeiro, apenas uma das partes precisa ser verdadeira.

Se o Python lê a primeira parte de um or e ela é True, ele faz o curto-circuito e ignora o resto, pois já sabe que o resultado final será True.

No and: Para de olhar no primeiro False.

No or: Para de olhar no primeiro True.

#### 'and' e 'or' retornam objetos

Eles não retornam obrigatoriamente True ou False.
```
print("" or "valor padrão")
```
Resultado: valor padrão
```
print("Gustavo" and 42)
```
Resultado: 42

Regras simplificadas:

a `or` b\
retorna a se a for verdadeiro;\
caso contrário, retorna b.\
a `and` b\
retorna a se a for falso;\
caso contrário, retorna b.

Isso permite:
```
nome_exibicao = apelido or nome_completo
```
Mas cuidado:
```
quantidade = quantidade_informada or 10
```
Se quantidade_informada for 0, o resultado será 10.

### `==` versus `is`

- `==` compara valores.
- `is` compara identidade: verifica se as variáveis apontam para o mesmo objeto.
- `is` principalmente para **singletons**, especialmente `None`.

````python
a = [1, 2]
b = [1, 2]

print(a == b)
print(a is b)
````

Uso correto de `is`:

### O que é um **Singleton**?

No contexto do Python, um Singleton é um objeto que só possui uma única cópia (instância) na memória durante toda a execução do programa.

Quando a documentação ou um tutorial diz que "o operador `is` é usado principalmente para singletons, \
especialmente `None`, isso faz todo o sentido devido à forma como o Python gerencia a sua memória interna.

Vamos destrinchar isso relacionando diretamente com as comparações `==` e `is`.

1. Por que `None` é um Singleton?
No Python, o `None` representa a ausência de valor. \
Para não desperdiçar memória criando um novo objeto "vazio" toda vez que você escreve `None`,\
 o Python cria apenas um objeto `None` na memória quando o programa inicializa.

Não importa em quantas variáveis diferentes você atribua `None`:\
todas elas apontarão exatamente para o mesmo endereço de memória. \
Ele é o Singleton mais famoso do Python.

2. A diferença crucial: `==` vs `is`
Para entender o papel do `is` com singletons, precisamos fixar a diferença visualmente:

`==` (Comparação de Valor): \
Pergunta "O que está dentro dessas caixas é igual?" (Chama o método interno __eq__).

`is` (Comparação de Identidade/Endereço): \
Pergunta "Essas duas variáveis apontam para a mesma caixa física na memória?" (Compara os IDs retornados pela função id()).

### O perigo de usar `==` com None
Se você usar `==`, um objeto malicioso ou mal projetado pode "mentir" para você.\
O operador `==` pode ser customizado dentro de uma classe alterando o método especial __eq__.


### Outros Singletons embutidos no Python

Além do None, existem outros objetos em Python que funcionam como Singletons (ou quase isso) e que também se deve comparar usando `is`:

`True` e `False`: Só existe um `True` e um `False` na memória. Escrever ```if ativo is True:```  funciona, embora o pythônico seja apenas ```if ativo:```.

Ellipsis (escrito como `...`): Usado em fatiamento de `arrays` e tipagem.

Pequenos inteiros (de `-5` a `256`): Para otimizar performance, o Python pré-carrega esses números na inicialização. Eles funcionam como "singletons temporários".


````python
x = 10
y = 10
print(x is y)  # True (apontam para o mesmo número pré-carregado)
print(x == y)

a = 3000
b = 3000
print(a is b)  # False (números grandes ganham novos endereços na memória)
print(a == b)
````

### Resumo prático
Use `==` quando você quiser saber se o conteúdo de duas variáveis é equivalente \
(ex: se dois textos são iguais, se duas listas têm os mesmos itens).

Use `is` quando estiver lidando com Singletons (especialmente `is None` ou `is not None`),\
pois você quer garantir de forma rápida e segura que a variável está apontando para aquela exata e única referência na memória.

### Condicionais aninhadas

O código funciona, mas produz uma “pirâmide” de indentação.

### Guard clauses
Profissionais frequentemente usam retornos antecipados:
```
def executar_operacao(usuario, recurso):
    if not usuario.ativo:
        return "Usuário inativo"

    if not usuario.possui_permissao:
        return "Permissão negada"

    if not recurso.disponivel:
        return "Recurso indisponível"

    recurso.executar()
    return "Operação concluída"
```
Vantagens:

reduz aninhamento;\
deixa pré-condições explícitas;\
mantém o fluxo principal alinhado à esquerda;\
facilita testes;\
reduz carga mental.

### Combinar ou separar condições?

Pode-se escrever:
```
if usuario.ativo and usuario.possui_permissao and recurso.disponivel:
    executar()
```
Isso é bom quando todas as condições têm:

o mesmo tratamento;\
a mesma mensagem;\
a mesma consequência.

Separe quando cada falha exige uma ação diferente:
```
if not usuario.ativo:
    raise UsuarioInativoError()

if not usuario.possui_permissao:
    raise PermissionError()

if not recurso.disponivel:
    raise RecursoIndisponivelError()
```
### Expressão condicional

```
resultado = valor_se_verdadeiro if condicao else valor_se_falso
```
Exemplo:
```

```

### Operador walrus `:=`
Ele recebe esse nome simpático ("walrus" significa morsa em inglês)\
porque o símbolo `:=` se parece com os olhos e as presas de uma morsa deitada de lado.
#### O Conceito Central: "Atribuir e Avaliar"

Normalmente, no Python, a atribuição (=) e a avaliação/comparação \
(if, while, etc.) são duas etapas completamente separadas.

O = é uma instrução (ele faz algo: guarda o valor na gaveta). \
Ele não retorna um valor diretamente. Por isso, você não pode fazer if x = 5: (dá erro de sintaxe).

O := é uma expressão de atribuição. Ele faz duas coisas ao mesmo tempo:

1. Guarda o valor na variável (assim como o =).

2. Retorna esse mesmo valor para que o Python possa usá-lo imediatamente na condição.

#### Analisando os Exemplos Práticos
1. Evitando repetição de código (len(texto))
Veja a comparação do seu exemplo:

- Sem o Walrus:
```
tamanho = len(texto) # Etapa 1: Atribui
if tamanho > 10:     # Etapa 2: Compara
    print(f"Texto longo: {tamanho} caracteres")
```
Aqui, a variável `tamanho` é criada na linha de cima apenas para ser usada no `if` e no `print`.

- Com o Walrus:
```
if (tamanho := len(texto)) > 10:
    print(f"Texto longo: {tamanho} caracteres")
```
O que o Python faz aqui?

Ele calcula `len(texto)` (que é 15).

Atribui `15` à variável `tamanho`.

Substitui toda a expressão `(tamanho := len(texto))` pelo valor retornado (15).

Avalia se `15 > 10` (que é `True`).

Como deu `True`, ele entra no `if` e você pode usar a variável `tamanho` lá dentro!

2. Uso clássico em loops (Leitura de arquivos)
Esse é um dos melhores cenários para o Walrus. Imagine ler um arquivo linha por linha.

- Sem o Walrus (Forma antiga e repetitiva):
```
linha = arquivo.readline()  # Precisa ler a primeira linha antes de entrar no loop
while linha:
    processar(linha)
    linha = arquivo.readline()  # Precisa ler a próxima linha no fim do loop para não travar
```
Repare que você teve que escrever `arquivo.readline()` duas vezes no código.

- Com o Walrus (Muito mais limpo):
```
while (linha := arquivo.readline()):
    processar(linha)
```
  A cada rodada do loop, o Python lê a linha, guarda em linha e avalia se ela não está vazia (se estiver vazia, o while para).\
  É um código sem repetição e eu achei muito elegante.

3. Uso com expressões regulares (`re.search`)
   Expressões regulares são famosas por exigir que você:
   - faça a busca,
   - verifique se encontrou algo (se não é None)
   - e só então extraia o resultado.
```
import re

if correspondencia := re.search(r"\d+", "Pedido 157"):
    numero = correspondencia.group()
    print(numero)
```
Se a busca encontrar o padrão (no caso, números 157), \
o objeto de correspondência é guardado na variável correspondencia e\
o bloco do if é executado. Se não encontrar nada, `re.search` retorna `None`,\
o `if` avalia como falso e nada quebra.

#### Boas Práticas: Quando usar e quando evitar

##### Quando usar (O lado bom):
- Para evitar computação repetida: Se você precisa rodar uma função pesada\
 (como uma consulta ao banco de dados ou um cálculo matemático complexo)\
 e usar o resultado dela tanto no teste (`if`) quanto no corpo do bloco.

- Proximidade: Ele cria a variável exatamente no lugar onde ela começa a ser útil,\
o que ajuda a entender o escopo dela no código.

##### Quando EVITAR:
# NÃO FAÇA ISSO! É péssimo para ler e debugar.
```if (a := obter_a()) and (b := obter_b()) and (c := calcular(a, b)):
    ...
```
**Por que isso é ruim?**

Curto-circuito traiçoeiro: Se `obter_a()` retornar `False` (ou `None`), \
o Python vai parar a execução do and imediatamente ali. \
Isso significa que as variáveis `b` e `c` nunca serão criadas!\
Se você tentar usá-las mais abaixo no código, \
seu programa vai quebrar com um erro de variável não definida (`NameError`).

Legibilidade: Fica extremamente difícil ler a linha e entender o que está sendo testado e o que está sendo atribuído. O código vira uma "sopa de letrinhas".

Regra de ouro: Use o operador Walrus apenas quando ele simplificar a leitura do seu código, nunca quando ele o tornar um enigma complexo.
