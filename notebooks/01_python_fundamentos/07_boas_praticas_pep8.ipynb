# 📘 Python - Boas Práticas e PEP8

## 🎯 Objetivos
- Conhecer as convenções de estilo do Python (PEP8)
- Escrever código legível e manutenível
- Usar ferramentas de linting (flake8, pylint, black)
- Documentar código com docstrings

## 📝 Conteúdo

### 1. PEP8 - Guia de Estilo
- Indentação (4 espaços)
- Nomes de variáveis, funções, classes
- Linhas de até 79 caracteres
- Espaços em branco

### 2. Docstrings
- `"""Docstring"""`
- Formato Google, NumPy ou Sphinx

### 3. Ferramentas de Linting
- `flake8`
- `pylint`
- `black` (formatador automático)

### 4. Boas Práticas Gerais
- Escrever código claro
- Evitar repetição (DRY)
- Comentar quando necessário
- Nomes significativos

## ✅ Exercícios
1. Refatorar um código para seguir PEP8
2. Adicionar docstrings a funções
3. Rodar flake8 em um arquivo
