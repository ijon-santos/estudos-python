## 1. O que é Python?

Python é uma linguagem de programação de alto nível, interpretada, de tipagem dinâmica e com suporte a múltiplos paradigmas.

### Origem

A linguagem nasceu em 1989 como um projeto de hobby durante as férias de Natal do programador holandês Guido van Rossum.\
Trabalhando no Centro de Matemática e Computação da Holanda (CWI), ele queria criar uma linguagem sucessora à linguagem ABC que fosse tão simples, mas que também corrigisse suas limitações.\
<br>
🐍 Origem do Nome\
Diferente do que muitos pensam, o nome não tem relação com a cobra. Guido era um grande fã do grupo de comédia britânico Monty Python.\
Ele queria um nome curto, marcante e ligeiramente misterioso, o que acabou gerando várias referências bem-humoradas na documentação oficial até hoje.\
<br>
👑 O Ditador Benevolente\
Guido van Rossum liderou o desenvolvimento e a tomada de decisões do Python por décadas, recebendo o título de BDFL (Benevolent Dictator for Life — Ditador Benevolente para a Vida Toda).\
Em 2018, ele se afastou do cargo de liderança máxima, e a comunidade adotou um comitê de diretores eleitos para guiar o futuro da linguagem. Em 2020, ele foi contratado pela Microsoft.\
<br>
🔥 A Transição Dolorosa (Python 2 vs. Python 3)\
Em 2008, o Python 3 foi lançado para corrigir falhas e problemas estruturais da linguagem.\
O grande porém é que ele não era compatível com o Python 2, forçando toda a comunidade a reescrever códigos antigos.\
Isso dividiu a comunidade de programadores por mais de 10 anos, até que o suporte oficial ao Python 2 fosse finalmente encerrado em 2020.\
<br>
🧘 O "Zen do Python"\
A linguagem é regida por uma filosofia que prioriza legibilidade e simplicidade, conhecida como "The Zen of Python".\
Trata-se de uma coleção de 19 princípios que ditam que "deve haver uma — e preferencialmente apenas uma — maneira óbvia de fazer as coisas".\
Você pode ler esse manifesto digitando "import this" em qualquer terminal Python.\
<br>
🚀 Dominância no Mercado\
Hoje, o Python é utilizado por gigantes como Google, Netflix, Spotify, Instagram, Dropbox e até pela NASA (para controle de robôs em missões).\
Com o avanço da Inteligência Artificial, Machine Learning e Ciência de Dados, a linguagem se tornou o pilar principal de bibliotecas poderosas como Pandas, NumPy e TensorFlow.

### Tipo de linguagem
- **Interpretada** → o código Python é executado linha a linha por **interpretadores**
- **Alta abstração / alto nível** → sintaxe próxima da linguagem humana, abstraindo gerenciamento de memória e hardware.
- **Multiparadigma** → suporta:
    - **Imperativa / procedural**
    - **Orientada a objetos (OOP)**
    - **Funcional** (funções de primeira classe, `map`, `filter`, `reduce`)
    - **Declarativa** (list comprehensions, expressões geradoras)

### Interpretadores Python

| Interpretador | Descrição | Quando usar |
|---------------|-----------|-------------|
| **CPython** | Implementação padrão, escrita em C | Uso geral, máximo de compatibilidade |
| **PyPy** | Implementação com JIT (Just-In-Time compilation) | Código com loops pesados, performance crítica |
| **Jython** | Implementação em Java, roda na JVM | Integração com código Java |
| **IronPython** | Implementação em C#, roda no .NET | Integração com código .NET |


### 🌟 Tabela Comparativa Avançada de Interpretadores Python

| Categoria | Interpretador / Implementação | Engine / Arquitetura Base | Compatibilidade com Ecossistema (`C-Extensions` / `PyPI`) | Gestão de Concorrência (Status do GIL) | Cenário de Uso Recomendado (*When to Use*) | Principais Limitações / *Trade-offs* |
| --- | --- | --- | --- | --- | --- | --- |
| **Padrão & Referência** | **CPython** *(v3.13 / v3.14+)* | Máquina Virtual de Stack (Bytecode) com compilador AOT invisível. | **100% (Referência Nativa)**. Suporte total a NumPy, SciPy, Pandas, etc. | **GIL Opcional (PEP 703)**. Suporte a *free-threading* e subinterpretadores isolados per-thread (PEP 684). | Aplicações generalistas, IA/Data Science nativa, ambientes de produção padrão e microsserviços conteinerizados. | Desempenho bruto inferior em loops puros de CPU em comparação com abordagens JIT avançadas. |
| **Alta Performance (JIT)** | **PyPy** | Tracing JIT (Just-In-Time) construído via framework RPython. | Alta através de emulação (`PyPy cpyext`), mas pode apresentar gargalos de performance com extensões C complexas. | **GIL Ativo**. Possui implementações internas de STM (Software Transactional Memory) experimentais. | Scripts de automação pesados, processamento de logs, algoritmos puros (CPU-bound) com execução prolongada (*warm-up*). | Alto consumo inicial de memória RAM; tempo de *warm-up* inicial elevado; incompatibilidade latente com certas libs C binárias. |
| **Ecosistema Java** | **Jython** | Compilação direta para Java Bytecode executado na JVM. | **Baixa**. Incompatível com a maioria das C-Extensions modernas do PyPI. Preso legadamente ao Python 2.7. | **Sem GIL**. Utiliza o modelo de concorrência nativo e multi-threading real da JVM. | Integração profunda em arquiteturas corporativas Java, automação de servidores WebSphere/WebLogic, reutilização de pacotes `.jar`. | Descontinuado ou sem suporte ativo para o ecossistema moderno do Python 3.x. |
| **Ecosistema .NET** | **IronPython** | Compilação para Microsoft IL (Intermediate Language) via DLR (Dynamic Language Runtime). | **Baixa**. Foco restrito a bibliotecas puras em Python e interoperabilidade com montagens .NET. | **Sem GIL**. Delega a gestão de concorrência às threads nativas do CLR (.NET). | Scripting para jogos baseados em motores .NET, automação de ferramentas CAD/Windows corporativas, extensão de apps C#/.NET. | Suporte a Python 3.x ainda instável/parcial; ecossistema severamente isolado do PyPI tradicional. |
| **Poliglota / Enterprise** | **GraalPy** *(Oracle)* | Compilação JIT baseada no framework GraalVM (Truffle AST). | Média-Alta. Excelente otimização e emulação nativa para C-Extensions em execução poliglota. | **GIL Adaptativo**. Gerido pelo ecossistema poliglota Graal. | Ambientes Enterprise que exigem execução poliglota de latência ultra-baixa (ex: chamar Python de dentro do Java/NodeJS sem overhead de IPC). | Requer infraestrutura GraalVM; pegada de memória substancial na inicialização da JVM. |
| **Embarcados / IoT** | **MicroPython** | Engine enxuta escrita em C, otimizada para execução direta em bare-metal. | **Nula (Ecossistema Próprio)**. Usa subset focado de bibliotecas otimizadas (`ujson`, `uasyncio`). | **Sem GIL tradicional**. Suporte a concorrência assíncrona orientada a eventos e interrupções de hardware. | Sistemas embarcados com restrição severa de hardware, microcontroladores (ESP32, STM32, RP2040), Internet das Coisas (IoT). | Subset limitado da sintaxe e tipos do Python; memória heap extremamente restrita (frequentemente < 256KB). |
| **Embarcados (Educação)** | **CircuitPython** | Fork do MicroPython mantido pela Adafruit. | **Nula**. Foco total em drivers de sensores e atuadores da Adafruit. | Single-thread / Loop principal orientado a hardware. | Aprendizado de eletrónica, prototipagem rápida de hardware, projetos Maker e robótica educacional para iniciantes. | Sem suporte a multi-threading avançado ou assincronismo complexo nativo do MicroPython Core. |
| **Híbrido de Compilação** | **Cython** *(Compilador/Runtime)* | Tradução de código Python/C-like em código fonte C/C++ altamente otimizado. | **100% (É a base de criação das extensões do PyPI)**. | **Libertação do GIL**. Permite o uso do bloco `with nogil` para paralelismo real em threads C. | Escrita de módulos de alta performance para o CPython, bindings de bibliotecas C/C++ legadas, otimização matemática crítica. | Curva de aprendizagem íngreme (exige conceitos de tipagem estática e gestão de memória C). |
| **IA de Próxima Geração** | **Mojo** *(Modular)* | Compilação AOT/JIT via infraestrutura MLIR/LLVM. Projetada como superconjunto do Python. | Alta (Interoperabilidade nativa através do mecanismo `Python.import_module`). | **Sem GIL**. Paralelismo nativo adaptado a arquiteturas massivas de CPU/GPU. | Desenvolvimento de modelos de Machine Learning de raiz, computação de alto desempenho (HPC), substituição de pipelines complexos C++/Python. | Linguagem em desenvolvimento ativo (ainda a consolidar o ecossistema completo fora do ecossistema de IA). |

---

1. **Atualização CPython JIT (PEP 744):** Com as recentes atualizações do CPython (focando no projeto *Faster CPython* das versões 3.13 e 3.14/3.15), o interpretador padrão agora possui um compilador JIT experimental integrado (*Copy-and-patch*). A necessidade de migrar para o PyPy apenas por performance bruta diminuiu para códigos genéricos.
2. **Arquitetura Multi-Thread Real (PEP 703 / 684):** O CPython introduziu suporte a múltiplos interpretadores isolados e a desativação opcional do GIL (*Free-Threading*), mudando drasticamente o paradigma de "Quando usar".
3. **Status de Compatibilidade de Ecossistema (`C-Extensions`):** O grande calcanhar de Aquiles dos interpretadores alternativos sempre foi quebrar bibliotecas como *NumPy* ou *Cryptography*. A coluna dedicada a isso previne arquitetos de software de escolherem um interpretador (como Jython) e descobrirem tarde demais que ele não instala dependências modernas do PyPI.
4. **Introdução do GraalPy e Mojo:** O GraalPy tornou-se a escolha corporativa para ambientes corporativos Java, e o Mojo foi incluído devido à sua proposta única de fusão de sintaxe Python com performance nativa de metal (LLVM/MLIR).

### Tipagem

- **Tipagem dinâmica** → o tipo das variáveis é inferido em tempo de execução.
- **Tipagem forte** → Python não realiza coerção automática de tipos incompatíveis (`'1' + 1` gera erro).
- **Suporte a anotações de tipo** → `x: int = 10`, útil para **tipagem estática com mypy**.


### Como o Python executa o código:

```
Código Fonte (.py) → Compilador → Bytecode (.pyc) → Interpretador → Execução
```

```python
# Exemplo: verificar qual interpretador está sendo usado

print(f"Versão: {sys.version}")

````python
import sys
print(f"Interpretador: {sys.implementation.name}")
print(f"Versão: {sys.version}")
````

### 2.2 Variáveis e Tipos de Dados

Python é **dinamicamente tipado** → não precisa declarar o tipo da variável.

| Tipo | Exemplo | Descrição |
|------|---------|-----------|
| `int` | `idade = 25` | Números inteiros |
| `float` | `altura = 1.75` | Números de ponto flutuante |
| `str` | `nome = "João"` | Cadeias de caracteres |
| `bool` | `True / False` | Valores lógicos |
| `list` | `[1, 2, 3]` | Lista (mutável) |
| `tuple` | `(1, 2, 3)` | Tupla (imutável) |
| `dict` | `{"nome": "João"}` | Dicionário (chave-valor) |
| `set` | `{1, 2, 3}` | Conjunto (não ordenado, sem duplicatas) |

````python
# Variáveis e tipos
idade = 25              # int
altura = 1.75           # float
nome = "João"           # str
estudando = True        # bool
frutas = ["maçã", "banana", "laranja"]  # list
pessoa = {"nome": "João", "idade": 25}   # dict

print(f"Nome: {nome}")
print(f"Idade: {idade}")
print(f"Altura: {altura}")
print(f"Estudando: {estudando}")
print(f"Frutas: {frutas}")
print(f"Pessoa: {pessoa}")
````

### 2.3 Verificando o Tipo de uma Variável

Use a função `type()` para descobrir o tipo de uma variável.

````python
print(type(idade))
print(type(nome))
print(type(frutas))
````
