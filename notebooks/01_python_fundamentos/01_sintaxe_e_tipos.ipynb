Em Python, existem poucas formas oficiais de comentar código, mas há várias técnicas usadas na prática. Abaixo está uma visão completa, classificando o que é comentário, documentação e outros recursos relacionados.

#### 2.1.5 Triple Quotes usadas como "comentário"

Muito comum:

````python
"""
Comentário
de várias linhas.
"""
````

ou

````python
'''
Comentário
'''
````

Mas isso NÃO é um comentário.

É apenas uma string.

Se ela aparecer sozinha:

````python
"""
teste
"""
````

o Python faz algo parecido com:

````python
temp = "teste"
````

e depois descarta.

Ou seja:

* cria um objeto string;
* pode ocupar memória (mesmo que brevemente);
* é executada.

Por isso, a Python Software Foundation e o guia de estilo recomendam não utilizar aspas triplas como substituto de comentários.

Boas práticas
* Use # para comentários.
* Use docstrings apenas para documentar módulos, classes, funções e métodos.
* Evite usar strings de aspas triplas como "comentários de bloco", pois elas não são comentários e ainda são processadas pelo interpretador.
* Aproveite marcadores como TODO, FIXME e NOTE quando quiser deixar tarefas ou observações para você ou para a equipe, especialmente porque muitas IDEs os destacam automaticamente.

### 2.2 Variáveis, Objetos e Tipos de Dados

Python é uma linguagem **fortemente tipada** (*Strongly Typed*) e **dinamicamente tipada** (*Dynamically Typed*).

- **Dinamicamente tipada:** o tipo é associado ao **objeto** em tempo de execução, e não à variável.
- **Fortemente tipada:** operações entre tipos incompatíveis não são realizadas automaticamente, evitando conversões implícitas que podem causar erros.

> **Importante**
>
> Em Python, **as variáveis não armazenam valores diretamente**. Elas armazenam **referências para objetos** na memória. Portanto, quem possui um tipo é o **objeto**, e não a variável.

#### Exemplo

```python
idade = 25
print(type(idade))

idade = "25"
print(type(idade))
```

**Saída:**

```python
<class 'int'>
<class 'str'>
```

Observe que a variável `idade` continua existindo, mas agora referencia um objeto diferente.

## Obtendo o Tipo de um Objeto

Utilize a função `type()` para descobrir o tipo de um objeto.

```python
nome = "Maria"
idade = 30
altura = 1.72

print(type(nome))
print(type(idade))
print(type(altura))
```

**Saída:**

```python
<class 'str'>
<class 'int'>
<class 'float'>
```

---

## Verificando o Tipo de um Objeto

Para verificar se um objeto pertence a um determinado tipo, utilize `isinstance()`.

```python
idade = 30

print(isinstance(idade, int))
print(isinstance(idade, float))
```

**Saída:**

```python
True
False
```

O uso de `isinstance()` é preferível ao de `type()` quando o objetivo é validar tipos, pois também considera herança.

## Boas Práticas

- Utilize nomes de variáveis claros, descritivos e significativos.
- Siga a convenção **snake_case** para nomes de variáveis.
- Evite reutilizar a mesma variável para armazenar objetos de tipos diferentes, exceto quando isso for realmente necessário.
- <span style="color:#FFD700; font-weight:bold;">Prefira `isinstance()` para verificar tipos em vez de comparar diretamente o resultado de `type()`.</span>
- Evite ocultar funções internas (*built-ins*) utilizando nomes como `list`, `dict`, `str`, `int`, `set`, `tuple`, `type`, `id` ou `input`.
- <span style="color:#FFD700; font-weight:bold;">Utilize **type hints** sempre que possível para documentar o tipo esperado de parâmetros, variáveis e retornos.</span> 
- <span style="color:#FFD700; font-weight:bold;">Lembre-se de que **quem possui tipo é o objeto, e não a variável**.</span>

Type hint é uma dica de tipo usada para tornar o código mais legível e facilitar a detecção de erros.

Exemplo rápido:

python
def soma(a: int, b: int) -> int:
    return a + b
Aqui, os hints dizem que a e b devem ser inteiros e que a função retorna um inteiro.

## Erros Comuns

❌ Acreditar que a variável possui um tipo fixo.

```python
x = 10
x = "dez"
```

A variável continua sendo a mesma; apenas passa a referenciar outro objeto.

---

❌ Comparar tipos usando `type()` quando há herança envolvida.

```python
type(obj) == MinhaClasse
```

✔️ Prefira:

```python
isinstance(obj, MinhaClasse)
```

---

❌ Utilizar nomes de variáveis iguais aos tipos nativos.

```python
list = [1, 2, 3]
```

Isso sobrescreve a referência à função/construtor `list()`, podendo causar erros posteriormente.

#### O que o input() realmente retorna

O input() sempre retorna uma str.

````python
idade = input("Idade: ")

print(type(idade))
````

Por isso:

````python
idade = int(input("Idade: "))
````

Essa diferença importa bastante quando começamos a tratar erros.

#### O problema de converter diretamente

Isto é comum:

##### idade = int(input("Idade: "))

Mas, se o usuário digitar: vinte

o programa quebra:ValueError

Um padrão mais robusto é:

````python
while True:
    try:
        idade = int(input("Digite sua idade: "))
        break
    except ValueError:
        print("Digite um número inteiro válido.")

print(f"Idade registrada: {idade}")
````

#### Criando uma função reutilizável de entrada

Em vez de repetir try/except:

````python
def ler_inteiro(mensagem):
    while True:
        try:
            return int(input(mensagem))
        except ValueError:
            print("Entrada inválida. Digite um número inteiro.")


idade = ler_inteiro("Idade: ")
quantidade = ler_inteiro("Quantidade: ")
````

Observe uma coisa interessante:

    return int(input(mensagem))

Quando o valor for válido, o return:

devolve o valor;
encerra a função;
consequentemente encerra o while.

Por isso não precisamos de break.

#### Validação com limites

Uma entrada numérica pode ser válida sintaticamente, mas inválida para sua regra de negócio.

Por exemplo: idade = -500

é um inteiro válido, mas provavelmente uma idade inválida.

Uma versão genérica:

````python
def ler_inteiro(mensagem, minimo=None, maximo=None):
    while True:
        try:
            valor = int(input(mensagem))

            if minimo is not None and valor < minimo:
                print(f"O valor mínimo é {minimo}.")
                continue

            if maximo is not None and valor > maximo:
                print(f"O valor máximo é {maximo}.")
                continue

            return valor

        except ValueError:
            print("Digite um número inteiro válido.")
````

map() é uma função embutida (built-in) do Python que aplica uma função a cada elemento de um iterável (lista, tupla, etc.) e retorna um novo iterador com os resultados.

````python
map(função, iterável)
````

````python
numeros = [1, 2, 3, 4, 5]

def dobrar(x):
    return x * 2

resultado = map(dobrar, numeros)
print(list(resultado))  # [2, 4, 6, 8, 10]
````

⚠️ Detalhe importante: map() retorna um objeto map (um iterador lazy), não uma lista diretamente. Por isso, geralmente convertemos com list(), tuple(), etc.

````python
resultado = map(dobrar, numeros)
print(resultado)          # <map object at 0x...>
print(list(resultado))    # [2, 4, 6, 8, 10]
````

Muita gente no Python moderno prefere list comprehension ao invés de map(), porque é mais legível:

````python
# Com map + lambda
resultado = list(map(lambda x: x * 2, numeros))

# Com list comprehension (geralmente preferido)
resultado = [x * 2 for x in numeros]
````

Ambos fazem a mesma coisa. map() costuma ser levemente mais rápido em alguns casos (principalmente quando usado com uma função nomeada, sem lambda), mas list comprehension é considerado mais "pythônico" (idiomático) pela comunidade.

### 3.3 `type()` ou `isinstance()` – Descobrir tipo
```python
type(variavel)
isinstance(valor,tipo)
```

````python
print(type(10))
print(type(10.5))
print(type("texto"))
print(isinstance(10, int))
print(isinstance(10.5, float))
print(isinstance("texto", str))
````

#### Por que isinstance() é recomendado?

Porque ele considera herança.

Exemplo:

````python
class Animal:
    pass

class Cachorro(Animal):
    pass

dog = Cachorro()

print(type(dog) == Animal)
print(isinstance(dog, Animal))
````

Embora dog seja uma instância de Cachorro, ele também é uma instância de Animal, já que Cachorro herda de Animal. type() verifica apenas o tipo exato, enquanto isinstance() respeita a hierarquia de herança.
